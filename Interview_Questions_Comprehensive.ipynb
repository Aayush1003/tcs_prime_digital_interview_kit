### Q2: What is a process and thread?

**Answer:**

**Process**
- Independent program execution
- Own memory space (code, data, stack)
- Heavy-weight (slower to create)
- Isolated from other processes
- Example: Chrome browser window

**Thread**
- Light-weight execution unit within process
- Shares process's memory
- Light-weight (faster to create)
- Share heap memory (access to globals)
- Example: Rendering thread, network thread in browser

**Comparison:**

| Aspect | Process | Thread |
|--------|---------|--------|
| Memory | Separate | Shared |
| Creation | Slow | Fast |
| Context Switch | Expensive | Cheap |
| Communication | Hard (IPC needed) | Easy (shared memory) |
| Overhead | High | Low |
| Isolation | Complete | None |
| Safety | Safer (isolated) | Risky (shared memory) |

**Process vs Thread Visualization:**
```
Process 1: [Memory] [Thread 1] → Code, Stack
                    [Thread 2] → Stack only
                    (Share heap memory)

Process 2: [Memory] [Thread 1] → Code, Stack
           (Isolated from Process 1)
```

### Q3: Explain CPU scheduling algorithms.

**Answer:**

**CPU Scheduling** - Decides which process gets CPU time next.

**FCFS (First-Come First-Served)**
```
Process: P1(24ms) → P2(3ms) → P3(3ms)
Execution: P1[0-24] P2[24-27] P3[27-30]
Average wait time = (0 + 24 + 27)/3 = 17ms

Problem: Short processes wait long (Convoy effect)
```

**SJF (Shortest Job First)**
```
Process: P1(24ms), P2(3ms), P3(3ms)
Execution: P2[0-3] P3[3-6] P1[6-30]
Average wait time = (3 + 6 + 0)/3 = 3ms  (Better!)

Problem: Long processes starved, not preemptive
```

**Round Robin (Preemptive)**
```
Time slice (quantum) = 10ms
Process: P1(24) P2(3) P3(3)

Execution:
[0-10]: P1 (14 remaining)
[10-13]: P2 (0 done)
[13-16]: P3 (0 done)
[16-26]: P1 (4 remaining)
[26-30]: P1 (0 done)

Fair, responsive, but more overhead
```

**Priority Scheduling**
```
Each process has priority
Higher priority runs first

Problem: Low priority starvation
Solution: Aging (increase priority over time)
```

**Multilevel Queue**
- Different queues for different priority levels
- System processes, interactive, batch, etc.

### Q4: Explain virtual memory and paging.

**Answer:**

**Virtual Memory** - Illusion of unlimited RAM using disk.

**Why Needed:**
- Physical RAM limited
- Multiple processes need more memory than available
- Allow processes larger than RAM

**How It Works:**
```
Virtual Address Space (32-bit: 4GB, 64-bit: 16EB)
       ↓ (Memory Management Unit - MMU)
Physical Memory (actual RAM, limited)

Process thinks it has 4GB
Actually might have only from physical RAM
Rest paged to disk as needed
```

**Paging:**
- Virtual memory divided into fixed-size pages (4KB typical)
- Physical memory divided into frames
- Page table maps virtual pages → physical frames

```
Virtual Address: [Page Number | Offset]
                 [0-19 bits   | 12 bits]

Page Table shows:
Page 0 → Frame 5
Page 1 → Disk (not in memory)
Page 2 → Frame 2
...

To access address:
1. Extract page number
2. Look up in page table
3. If in memory: calculate physical address
4. If on disk: page fault → load from disk
```

**Page Fault:**
```
1. Requested page not in memory
2. OS loads page from disk
3. Expensive operation (milliseconds)
4. Resume process

Strategy: Minimize page faults
- LRU (Least Recently Used) replacement
- FIFO, Clock algorithm
```

**Advantages:**
- Processes appear larger than RAM
- Memory protection (each process isolated)
- Easier programming (don't worry about RAM size)

**Disadvantages:**
- Page faults cause delays
- Disk I/O slower than RAM (100-1000x)
- Overhead of page table

### Q8: Explain process synchronization and deadlock.

**Answer:**

**Synchronization Problem:**
```
Two threads access shared variable:
count = 0

Thread 1:                Thread 2:
read count (0)           read count (0)
add 1                    add 1
write count (1)          write count (1)

Expected: count = 2
Actual: count = 1  (Lost update!)
```

**Critical Section Solution:**

```python
import threading

# Mutex lock
lock = threading.Lock()
count = 0

def increment():
    global count
    with lock:  # Acquire lock
        # Critical section (only one thread at a time)
        temp = count
        temp += 1
        count = temp
        # Release lock

thread1 = threading.Thread(target=increment)
thread2 = threading.Thread(target=increment)
thread1.start()
thread2.start()
# count = 2 (correct)
```

**Semaphore:**
```
Binary semaphore = lock (0 or 1)
Counting semaphore = counter (multiple resources)

P(s): Decrement (wait if 0)
V(s): Increment (signal)
```

**Deadlock** - Processes waiting for resources held by each other.

```
Process A: Has resource X
           Wants resource Y
           
Process B: Has resource Y
           Wants resource X
           
→ Both wait forever (DEADLOCK)
```

**Conditions for Deadlock:**
1. Mutual exclusion (resource not shareable)
2. Hold and wait (hold while waiting)
3. No preemption (can't forcibly take resource)
4. Circular wait (P1→P2→P3→P1)

**Prevention Strategies:**
```
1. Break mutual exclusion: Share resources
2. Break hold and wait: Request all resources upfront
3. Break no preemption: Allow resource preemption
4. Break circular wait: Resource ordering
```

### Q12: Design a distributed cache system like Memcached.

**Answer:**

**Requirements:**
- High-speed access
- Distributed across servers
- Handle failures gracefully
- Consistent hashing

**Architecture:**

```
Client Code
    ↓
Cache Client Library (consistent hashing)
    ↓
[Cache Server 1] [Cache Server 2] [Cache Server 3]
     (In-memory)      (In-memory)      (In-memory)
```

**Consistent Hashing (for scalability):**

```
Without consistent hashing (modulo):
key % num_servers = server_id
Problem: Adding server changes all keys!

With consistent hashing:
- Hash servers and keys on ring
- Key goes to next server clockwise
- Adding server: only affects nearby keys
```

**Implementation:**

```python
import hashlib

class ConsistentHash:
    def __init__(self, nodes=None, replicas=3):
        self.replicas = replicas
        self.ring = {}  # hash value → node
        self.sorted_keys = []
        
        if nodes:
            for node in nodes:
                self.add_node(node)
    
    def _hash(self, key):
        return int(hashlib.md5(key.encode()).hexdigest(), 16)
    
    def add_node(self, node):
        for i in range(self.replicas):
            virtual_key = f"{node}:{i}"
            hash_val = self._hash(virtual_key)
            self.ring[hash_val] = node
        
        self.sorted_keys = sorted(self.ring.keys())
    
    def get_node(self, key):
        if not self.ring:
            return None
        
        hash_val = self._hash(key)
        
        # Find first node >= key
        for ring_hash in self.sorted_keys:
            if hash_val <= ring_hash:
                return self.ring[ring_hash]
        
        # Wrap around
        return self.ring[self.sorted_keys[0]]
```

**Cache Operations:**

```python
class DistributedCache:
    def __init__(self, servers):
        self.hash_ring = ConsistentHash(servers)
        self.servers = {s: {} for s in servers}
    
    def get(self, key):
        server = self.hash_ring.get_node(key)
        if server and key in self.servers[server]:
            return self.servers[server][key]
        return None
    
    def set(self, key, value, ttl=None):
        server = self.hash_ring.get_node(key)
        self.servers[server][key] = value
        if ttl:
            # Schedule deletion after TTL
            pass
    
    def delete(self, key):
        server = self.hash_ring.get_node(key)
        if key in self.servers[server]:
            del self.servers[server][key]
```

**Replication for Fault Tolerance:**

```python
def set_with_replication(self, key, value, replication_factor=3):
    # Find N nodes starting from hash position
    server = self.hash_ring.get_node(key)
    replicated_servers = []
    
    for node in self.hash_ring.sorted_keys:
        if node >= server and len(replicated_servers) < replication_factor:
            replicated_servers.append(node)
    
    # Write to all replicas
    for server in replicated_servers:
        self.servers[server][key] = value
```

### Q14: Explain ACID properties and transactions.

**Answer:**

**ACID** ensures database reliability.

**1. Atomicity (All or Nothing)**
```
Transfer $100 from A to B:
1. Debit $100 from A
2. Credit $100 to B

If fail between steps:
→ Rollback both
→ Never left in between state
```

**2. Consistency (Valid to Valid)**
```
Before: A=500, B=300, total=800
Transfer: A pays B $100
After: A=400, B=400, total=800

Invariant maintained: total always 800
```

**3. Isolation (No Interference)**
```
Transaction 1: A=A-100
Transaction 2: B=B-100

If concurrent:
- T1 might see T2's intermediate state
- Dirty read, phantom read, non-repeatable read

Levels:
- Serializable: No conflicts
- Repeatable Read: Prevents non-repeatable read
- Read Committed: Prevents dirty read
- Read Uncommitted: Fastest, least safe
```

**4. Durability (Permanent)**
```
After commit, data persists
Even if power failure
Typically: Write-ahead logging (WAL)
- Write to log first
- Commit in log
- Then update database
```

**Transaction Control:**

```sql
BEGIN TRANSACTION;

UPDATE accounts SET balance = balance - 100
WHERE account_id = 1;

UPDATE accounts SET balance = balance + 100
WHERE account_id = 2;

-- If all OK:
COMMIT;

-- If error:
ROLLBACK;
```

### Q15: Design a messaging queue system (like RabbitMQ).

**Answer:**

**Requirements:**
- Asynchronous message passing
- Reliability (no message loss
- Scalability (many producers/consumers)
- Ordering (optional)

**Architecture:**

```
Producer 1 ┐
Producer 2 ├→ [Message Queue] → [Consumer 1]
Producer 3 ┘                  → [Consumer 2]
                              → [Consumer 3]
```

**Core Components:**

```python
from collections import deque
import threading

class MessageQueue:
    def __init__(self, max_size=1000):
        self.queue = deque(maxlen=max_size)
        self.lock = threading.Lock()
        self.not_empty = threading.Condition(self.lock)
    
    def publish(self, message):
        """Producer: Add message"""
        with self.not_empty:
            self.queue.append(message)
            self.not_empty.notify_all()  # Wake consumers
    
    def consume(self, timeout=None):
        """Consumer: Get message"""
        with self.not_empty:
            # Wait if empty
            while not self.queue:
                self.not_empty.wait(timeout)
            
            if self.queue:
                return self.queue.popleft()
            return None
    
    def size(self):
        with self.lock:
            return len(self.queue)
```

**Exchange & Routing:**

```
Message sent to Exchange
Exchange routes based on routing key

Types:
- Direct: Exact match
- Topic: Pattern matching (*.order.#)
- Fanout: Broadcast to all
- Headers: Match on headers
```

**Reliability - Acknowledgments:**

```
Consumer receives message
Process it
Send ACK back

If no ACK:
→ Message requeued
→ Other consumer tries
→ No message loss
```

**Persistence:**

```
Message written to disk before ACK
Even if broker crashes → recovered
Redis AOF (Append Only File)
or RabbitMQ's disk queues
```

**Implementation Pattern:**

```python
class MQBroker:
    def __init__(self):
        self.exchanges = {}
        self.queues = {}
    
    def declare_queue(self, name):
        self.queues[name] = MessageQueue()
    
    def publish(self, exchange, routing_key, message):
        # Route message based on exchange type
        target_queues = self._route(exchange, routing_key)
        
        for queue_name in target_queues:
            self.queues[queue_name].publish(message)
    
    def consume(self, queue_name, timeout=None):
        return self.queues[queue_name].consume(timeout)
```

**Scalability Considerations:**

1. **Sharding**: Partition by routing key
2. **Replication**: Replicate queue on backup nodes
3. **Load Balancing**: Distribute consumers
4. **Monitoring**: Track queue depth, latency

**Real-World Examples:**
- RabbitMQ: AMQP protocol
- Apache Kafka: Distributed streaming
- AWS SQS: Managed service
- Redis Streams: Simpler alternative

### Q8: Implement a stack and queue data structure.

**Answer:**

**Stack (LIFO - Last In First Out)**

```python
class Stack:
    def __init__(self):
        self.items = []
    
    def push(self, item):
        self.items.append(item)
    
    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)

# Operations: O(1) for push, pop, peek
# Applications: Function calls, undo/redo, expression evaluation
```

**Queue (FIFO - First In First Out)**

```python
from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()
    
    def enqueue(self, item):
        self.items.append(item)
    
    def dequeue(self):
        if not self.is_empty():
            return self.items.popleft()
        return None
    
    def peek(self):
        if not self.is_empty():
            return self.items[0]
        return None
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)

# Note: Use deque, not list (list.pop(0) is O(n))
# Operations: O(1) for enqueue, dequeue
# Applications: BFS, job scheduling, request queue
```

### Q9: Longest common subsequence (Dynamic Programming).

**Answer:**

**Problem:** Find longest subsequence common to two strings.

```
String 1: "abcde"
String 2: "ace"
LCS: "ace" (length 3)

String 1: "hello"
String 2: "world"
LCS: "lo" (length 2)
```

**DP Approach:**

```python
def lcs_length(text1, text2):
    m, n = len(text1), len(text2)
    
    # dp[i][j] = LCS length of text1[0:i] and text2[0:j]
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                # Characters match: add to previous LCS
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                # Characters don't match: take max from previous
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[m][n]

# Time: O(m × n), Space: O(m × n)
```

**Getting the Actual LCS (Backtrack from DP table):**

```python
def lcs_string(text1, text2):
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Build DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i - 1] == text2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    # Backtrack to find LCS string
    result = []
    i, j = m, n
    
    while i > 0 and j > 0:
        if text1[i - 1] == text2[j - 1]:
            result.append(text1[i - 1])
            i -= 1
            j -= 1
        elif dp[i - 1][j] > dp[i][j - 1]:
            i -= 1
        else:
            j -= 1
    
    return ''.join(reversed(result))

# Example
lcs_string("abcde", "ace")  # "ace"
```

### Q10: Implement binary search tree operations (insert, search, delete).

**Answer:**

**BST Structure:**

```python
class TreeNode:
    def __init__(self, val=0):
        self.val = val
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, val):
        if not self.root:
            self.root = TreeNode(val)
        else:
            self._insert_recursive(self.root, val)
    
    def _insert_recursive(self, node, val):
        if val < node.val:
            if not node.left:
                node.left = TreeNode(val)
            else:
                self._insert_recursive(node.left, val)
        else:
            if not node.right:
                node.right = TreeNode(val)
            else:
                self._insert_recursive(node.right, val)
    
    # Time: O(log n) average, O(n) worst (skewed tree)
    # Space: O(h) where h is height
```

**Search Operation:**

```python
def search(self, val):
    return self._search_recursive(self.root, val)

def _search_recursive(self, node, val):
    if not node:
        return False
    
    if node.val == val:
        return True
    elif val < node.val:
        return self._search_recursive(node.left, val)
    else:
        return self._search_recursive(node.right, val)

# Time: O(log n) average, O(n) worst
```

**Delete Operation (Complex):**

```python
def delete(self, val):
    self.root = self._delete_recursive(self.root, val)

def _delete_recursive(self, node, val):
    if not node:
        return None
    
    if val < node.val:
        node.left = self._delete_recursive(node.left, val)
    elif val > node.val:
        node.right = self._delete_recursive(node.right, val)
    else:
        # Found node to delete
        
        # Case 1: No children (leaf)
        if not node.left and not node.right:
            return None
        
        # Case 2: One child
        if not node.left:
            return node.right
        if not node.right:
            return node.left
        
        # Case 3: Two children
        # Find in-order successor (smallest in right subtree)
        min_right = self._find_min(node.right)
        node.val = min_right.val
        node.right = self._delete_recursive(node.right, min_right.val)
    
    return node

def _find_min(self, node):
    while node.left:
        node = node.left
    return node

# Time: O(log n) average, O(n) worst
```

### Q11: Implement graph traversal - BFS and DFS.

**Answer:**

**Graph Representation:**

```python
# Adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'E'],
    'D': ['B'],
    'E': ['C']
}
```

**BFS (Breadth-First Search):**

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    
    while queue:
        node = queue.popleft()
        
        if node not in visited:
            visited.add(node)
            result.append(node)
            
            # Add neighbors to queue
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    
    return result

# Time: O(V + E), Space: O(V)
# bfs(graph, 'A')  # ['A', 'B', 'C', 'D', 'E']
```

**DFS (Depth-First Search):**

```python
def dfs_recursive(graph, node, visited=None, result=None):
    if visited is None:
        visited = set()
        result = []
    
    visited.add(node)
    result.append(node)
    
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited, result)
    
    return result

# Time: O(V + E), Space: O(V) for visited set + call stack

def dfs_iterative(graph, start):
    visited = set()
    stack = [start]
    result = []
    
    while stack:
        node = stack.pop()
        
        if node not in visited:
            visited.add(node)
            result.append(node)
            
            # Add neighbors to stack (reverse order for same traversal)
            stack.extend(reversed(graph[node]))
    
    return result

# dfs_recursive(graph, 'A')  # ['A', 'B', 'D', 'C', 'E']
```

**Comparison:**

| Aspect | BFS | DFS |
|--------|-----|-----|
| Data Structure | Queue | Stack |
| Traversal | Level-by-level | Deep-first |
| Shortest Path | Unweighted graphs | No |
| Memory | More (all level) | Less |
| Applications | Shortest path, levels | Topological sort, cycles |

### Q12: Longest increasing subsequence (LIS) - Dynamic Programming.

**Answer:**

**Problem:** Find length of longest increasing subsequence.

```
Array: [10, 9, 2, 5, 3, 7, 101, 18]
LIS: [2, 3, 7, 101] or [2, 3, 7, 18]
Length: 4
```

**O(n²) DP Solution:**

```python
def lengthOfLIS(nums):
    if not nums:
        return 0
    
    n = len(nums)
    # dp[i] = length of LIS ending at index i
    dp = [1] * n
    
    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp)

# Time: O(n²), Space: O(n)
```

**O(n log n) Solution (Binary Search):**

```python
import bisect

def lengthOfLIS_optimized(nums):
    # tails[i] = smallest tail of all increasing subsequences of length i+1
    tails = []
    
    for num in nums:
        # Find position where num should be inserted
        pos = bisect.bisect_left(tails, num)
        
        if pos == len(tails):
            tails.append(num)
        else:
            tails[pos] = num
    
    return len(tails)

# Time: O(n log n), Space: O(n)

# Example trace:
# [10]:           tails = [10]
# [10, 9]:        tails = [9]         (9 replaces 10)
# [10, 9, 2]:     tails = [2]
# [10, 9, 2, 5]:  tails = [2, 5]
# [10, 9, 2, 5, 3]: tails = [2, 3]
# [10, 9, 2, 5, 3, 7]: tails = [2, 3, 7]
# Result: length = 3
```

### Q13: Word Ladder - BFS shortest path.

**Answer:**

**Problem:** Find shortest transformation from start to end word, changing one letter at a time.

```
Start: "hit"
End: "cog"
Word list: ["hot","dot","dog","lot","log","cog"]

Path: hit → hot → dot → dog → cog (length = 5)
```

**BFS Solution:**

```python
from collections import defaultdict, deque

def ladderLength(beginWord, endWord, wordList):
    if endWord not in wordList:
        return 0
    
    # Build adjacency graph for words one letter apart
    word_set = set(wordList)
    neighbors = defaultdict(list)
    
    def get_neighbors(word):
        result = []
        for i in range(len(word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                new_word = word[:i] + c + word[i+1:]
                if new_word in word_set:
                    result.append(new_word)
        return result
    
    # BFS
    queue = deque([(beginWord, 1)])
    visited = set([beginWord])
    
    while queue:
        word, length = queue.popleft()
        
        if word == endWord:
            return length
        
        for neighbor in get_neighbors(word):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, length + 1))
    
    return 0

# Time: O(N × L × 26) where N=word list size, L=word length
# Space: O(N × L)
```

**Bidirectional BFS (Optimization):**

```python
def ladderLength_bidirectional(beginWord, endWord, wordList):
    if endWord not in wordList:
        return 0
    
    word_set = set(wordList)
    
    def get_neighbors(word):
        result = []
        for i in range(len(word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                new_word = word[:i] + c + word[i+1:]
                if new_word in word_set and new_word!= word:
                    result.append(new_word)
        return result
    
    # Two BFS from both ends
    begin_set = {beginWord}
    end_set = {endWord}
    visited = {beginWord, endWord}
    length = 1
    
    while begin_set:
        length += 1
        
        # Expand smaller set (optimization)
        if len(begin_set) > len(end_set):
            begin_set, end_set = end_set, begin_set
        
        next_level = set()
        
        for word in begin_set:
            for neighbor in get_neighbors(word):
                if neighbor in end_set:
                    return length
                
                if neighbor not in visited:
                    visited.add(neighbor)
                    next_level.add(neighbor)
        
        begin_set = next_level
    
    return 0

# Time: O(N × L × 26), Space: O(N)
# Faster in practice due to smaller search space
```

### Q14: Implement LeetCode 460 - LFU Cache.

**Answer:**

**Problem:** Implement LFU Cache with get() and put() in O(1) time.
- Least Frequently Used item evicted when capacity exceeded
- If tie, least recently used among them

```python
from collections import defaultdict

class LFUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.key_to_val = {}      # key → value
        self.key_to_freq = {}      # key → frequency
        self.freq_to_keys = defaultdict(set)  # frequency → set of keys
        self.min_freq = 0
    
    def get(self, key):
        if key not in self.key_to_val:
            return -1
        
        self._update_frequency(key)
        return self.key_to_val[key]
    
    def put(self, key, value):
        if self.capacity <= 0:
            return
        
        if key in self.key_to_val:
            # Update existing key
            self.key_to_val[key] = value
            self._update_frequency(key)
        else:
            # Add new key
            if len(self.key_to_val) >= self.capacity:
                # Evict LFU
                self._evict()
            
            self.key_to_val[key] = value
            self.key_to_freq[key] = 1
            self.freq_to_keys[1].add(key)
            self.min_freq = 1
    
    def _update_frequency(self, key):
        freq = self.key_to_freq[key]
        
        # Remove from current frequency
        self.freq_to_keys[freq].discard(key)
        
        # If no keys at this frequency and it's min_freq, increment min_freq
        if not self.freq_to_keys[freq] and freq == self.min_freq:
            self.min_freq += 1
        
        # Add to new frequency
        new_freq = freq + 1
        self.key_to_freq[key] = new_freq
        self.freq_to_keys[new_freq].add(key)
    
    def _evict(self):
        # Remove oldest key from min_freq set
        oldest_key = next(iter(self.freq_to_keys[self.min_freq]))
        self.freq_to_keys[self.min_freq].remove(oldest_key)
        
        del self.key_to_val[oldest_key]
        del self.key_to_freq[oldest_key]

# Time: O(1) for both get and put
# Space: O(capacity)

# Usage:
# cache = LFUCache(2)
# cache.put(1, 1)
# cache.put(2, 2)
# cache.get(1)      # 1
# cache.put(3, 3)   # Evicts key 2
# cache.get(2)      # -1 (evicted)
```

### Q15: Implement Trie data structure with search and autocomplete.

**Answer:**

**Trie Structure:** Tree for storing strings, with nodes representing characters.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word):
        node = self.root
        
        for char in word:
            # Create node if doesn't exist
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        
        node.is_end_of_word = True
    
    # Time: O(m) where m = length of word
    # Space: O(m) in worst case
    
    def search(self, word):
        node = self._find_node(word)
        return node and node.is_end_of_word
    
    # Time: O(m)
    
    def starts_with(self, prefix):
        return self._find_node(prefix) is not None
    
    # Time: O(m)
    
    def _find_node(self, word):
        node = self.root
        
        for char in word:
            if char not in node.children:
                return None
            node = node.children[char]
        
        return node
    
    def autocomplete(self, prefix):
        """Find all words with given prefix"""
        node = self._find_node(prefix)
        if not node:
            return []
        
        results = []
        self._dfs(node, prefix, results)
        return results
    
    def _dfs(self, node, current_word, results):
        if node.is_end_of_word:
            results.append(current_word)
        
        for char, child_node in node.children.items():
            self._dfs(child_node, current_word + char, results)
    
    # Time: O(n) where n = number of nodes in subtree
```

**Usage Example:**

```python
trie = Trie()
trie.insert("apple")
trie.insert("application")
trie.insert("apply")
trie.insert("apricot")

print(trie.search("apple"))           # True
print(trie.search("app"))             # False
print(trie.starts_with("app"))        # True
print(trie.autocomplete("app"))       # ['apple', 'application', 'apply']
print(trie.autocomplete("apr"))       # ['apricot']
```

**Complexity:**
- Insert: O(m)
- Search: O(m)
- Autocomplete: O(n)
- Space: O(ALPHABET_SIZE × m × n) where n=number of words, m=avg word length

**Applications:**
- Spell checker
- IP routing
- Autocomplete
- Phone directory

### Q2: Check if a string is palindrome.

**Answer:**

**Two Pointer Approach:**
```python
def is_palindrome(s):
    # Keep only alphanumeric, convert lowercase
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    
    left, right = 0, len(cleaned) - 1
    
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Examples:
is_palindrome("A man, a plan, a canal: Panama")  # True
is_palindrome("race a car")                      # False
```

**With Space-Optimized Pointer Validation:**
```python
def is_palindrome_optimized(s):
    left, right = 0, len(s) - 1
    
    while left < right:
        # Skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        
        # Skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1
        
        # Compare (case-insensitive)
        if s[left].lower() != s[right].lower():
            return False
        
        left += 1
        right -= 1
    
    return True

# Time: O(n), Space: O(1)
```

### Q4: Linear search vs Binary search.

**Answer:**

**Linear Search**
```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Time: O(n), Space: O(1)
# Works on unsorted arrays
```

**Binary Search**
```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1  # Search right half
        else:
            right = mid - 1  # Search left half
    
    return -1

# Time: O(log n), Space: O(1)
# Requires SORTED array
```

**Comparison:**

| Aspect | Linear | Binary |
|--------|--------|--------|

| Time Complexity | O(n) | O(log n) |
| Implementation | Simple | Slightly complex |
| For array of 1 million | ~500K checks | ~20 checks |

**When to Use:**
- Linear: Small arrays or unsorted data
- Binary: Large sorted arrays, frequently searched

### Q5: Implementation of basic sorting - Bubble Sort.

**Answer:**

**Bubble Sort** - Repeatedly swap adjacent elements if in wrong order.

```python
def bubble_sort(arr):
    n = len(arr)
    
    # Outer loop: number of passes
    for i in range(n):
        # Flag to optimize
        swapped = False
        
        # Inner loop: compare adjacent elements
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        
        # If no swaps, array sorted
        if not swapped:
            break
    
    return arr

# Time: O(n²) average/worst, O(n) best (already sorted)
# Space: O(1) in-place
# Stable: Yes

# Example:
bubble_sort([64, 34, 25, 12, 22, 11, 90])
# [11, 12, 22, 25, 34, 64, 90]
```

**Trace through small example:**
```
[3, 1, 2]

Pass 1: Compare adjacent, bubble largest to end
3>1: [1, 3, 2]
3>2: [1, 2, 3]

Pass 2: Already sorted, swapped=False, break

Result: [1, 2, 3]
```

**Characteristics:**
- Simple to understand
- Inefficient for large arrays
- Rarely used in practice
- Good for learning, teaching

### Q6: Merge two sorted arrays.

**Answer:**

**Two Pointer Approach:**
```python
def merge_sorted_arrays(arr1, arr2):
    i, j = 0, 0
    result = []
    
    # Compare elements from both arrays
    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1
    
    # Append remaining elements
    result.extend(arr1[i:])
    result.extend(arr2[j:])
    
    return result

# Time: O(m + n), Space: O(m + n)
merge_sorted_arrays([1, 3, 5], [2, 4, 6])
# [1, 2, 3, 4, 5, 6]
```

**In-Place Merge (If space available at end of arr1):**
```python
def merge_sorted_arrays_inplace(arr1, m, arr2, n):
    # arr1 has space for n more elements
    i = m - 1
    j = n - 1
    k = m + n - 1
    
    # Merge from back (avoid overwriting)
    while i >= 0 and j >= 0:
        if arr1[i] > arr2[j]:
            arr1[k] = arr1[i]
            i -= 1
        else:
            arr1[k] = arr2[j]
            j -= 1
        k -= 1
    
    # Remaining arr2 elements
    while j >= 0:
        arr1[k] = arr2[j]
        j -= 1
        k -= 1
    
    return arr1

# Time: O(m + n), Space: O(1)
```

### Q7: Basic recursion - Calculate factorial.

**Answer:**

**Recursive Approach:**
```python
def factorial(n):
    # Base case
    if n == 0 or n == 1:
        return 1
    
    # Recursive case
    return n * factorial(n - 1)

# Examples:
factorial(5)  # 5 × 4 × 3 × 2 × 1 = 120
factorial(0)  # 1

# Time: O(n), Space: O(n) due to call stack
```

**Iterative Approach (More Efficient):**
```python
def factorial_iterative(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Time: O(n), Space: O(1)
```

**With Memoization (Prevent Recalculation):**
```python
def factorial_memo(n, memo={}):
    if n in memo:
        return memo[n]
    
    if n == 0 or n == 1:
        return 1
    
    memo[n] = n * factorial_memo(n - 1, memo)
    return memo[n]

# Time: O(n), Space: O(n)
# Avoids recalculation
```

**Common Issues with Recursion:**
- Stack overflow for large n
- Repeated calculations without memoization
- Each recursion call uses memory

**Best Practice:** Use iterative for factorial, but recursion is essential for tree/graph traversal.

### Q12: Explain generative models: VAE and GAN.

**Answer:**

**Generative Models** - Models that learn data distribution and can generate new samples.

**VAE (Variational Autoencoder)**

**Architecture:**
```
Encoder: Image → Latent space (compressed representation)
Decoder: Latent space → Reconstructed image

Key Difference from standard Autoencoder:
- Produces probability distribution (mean, variance)
- Sample from this distribution (adds randomness)
```

**How it Works:**
```
1. Encoder maps input to distribution in latent space
   Input: Face image → Mean=[1.2, 0.3], Variance=[0.5, 0.4]

2. Sample from distribution
   z ~ N(mean, variance)

3. Decoder reconstructs from z
   z → Reconstructed face (slightly different each time)
```

**Loss Function:**
```
L = Reconstruction Loss + KL Divergence
  = How well can we reconstruct + How close to standard normal
```

**Advantages:**
- Smooth latent space (can interpolate)
- Generate new realistic samples
- Interpretable latent dimensions

**Applications:**
- Image generation
- Data augmentation
- Anomaly detection

**GAN (Generative Adversarial Network)**

**Architecture - Two Networks Fighting:**
```
Generator: Random noise → Fake images
Discriminator: Images → Real or Fake?

Training:
- Generator tries to fool discriminator
- Discriminator tries to distinguish real vs fake
- Adversarial game → both improve
```

**How it Works:**
```
1. Generator creates fake image from random vector
2. Discriminator sees mix of real and fake
3. Penalizes generator for failed fakes
4. Penalizes discriminator for wrong classification
5. Both iteratively improve

Eventually:
- Generator makes indistinguishable fakes
- Discriminator at 50% accuracy (random guessing)
```

**Training Loop:**
```python
for epoch in range(num_epochs):
    # Train discriminator
    real_images = get_real_batch()
    fake_images = generator.generate()
    
    d_loss_real = discriminator_loss(real_images, real=True)
    d_loss_fake = discriminator_loss(fake_images, real=False)
    discriminator.update(d_loss_real + d_loss_fake)
    
    # Train generator
    fake_images = generator.generate()
    g_loss = discriminator_loss(fake_images, real=True)  # Fool discriminator
    generator.update(g_loss)
```

**Advantages:**
- Sharp, realistic images (often better than VAE)
- No encoder needed
- Can generate diverse outputs

**Disadvantages:**
- Harder to train (unstable)
- Mode collapse (generates limited variety)
- No latent space interpolation

**Comparison:**

| Aspect | VAE | GAN |
|--------|-----|-----|
| **Image Quality** | Blurry | Sharp |
| **Training Stability** | Stable | Unstable |
| **Latent Space** | Smooth | Not interpretable |
| **Diversity** | Good | Mode collapse risk |
| **Applications** | Reconstruction, VAE | High-quality synthesis |

**Applications:**
- Image generation (StyleGAN)
- Style transfer
- Data synthesis
- Super-resolution

### Q14: Explain advanced optimization: Learning Rate Scheduling, Batch Normalization, Dropout.

**Answer:**

**Learning Rate Scheduling** - Vary learning rate during training.

**Problem with Fixed Learning Rate:**
```
Too high: Jumps over optima, diverges
Too low: Slow convergence, gets stuck
Fixed: Can't adapt as training progresses
```

**Strategies:**

1. **Exponential Decay**
   ```python
   lr_t = lr_0 × 0.95^epoch
   
   # Start with high LR (explore), gradually decrease (fine-tune)
   ```

2. **Cosine Annealing**
   ```python
   lr_t = lr_min + 0.5 × (lr_max - lr_min) × (1 + cos(π × t/T))
   
   # Smoothly decreases, good performance
   ```

3. **Warm-up then Decay**
   ```python
   First N epochs: Gradually increase to lr_max
   After N epochs: Decay based on schedule
   
   # Stabilizes initial training
   ```

4. **Cyclical Learning Rate**
   ```python
   # Cycle between lr_min and lr_max
   # Helps escape local minima
   # Simulates annealing multiple times
   ```

**Batch Normalization (BatchNorm)**

**Problem it Solves:**
```
Different input scales hurt training
Layer 1 outputs: [0, 10]
Layer 2 assumes: [-1, 1]
Mismatch → slow training
```

**How it Works:**
```
For each batch:
1. Compute mean and variance
2. Normalize: (x - batch_mean) / batch_std
3. Scale and shift: y = γ × x_norm + β

Benefits:
- Stabilizes training
- Higher learning rates possible
- Reduces internal covariate shift
- Acts as regularization
```

**Code:**
```python
model = Sequential([
    Dense(128),
    BatchNormalization(),
    Activation('relu'),
    Dense(64),
    BatchNormalization(),
    Activation('relu'),
    Dense(10)
])
```

**Dropout**

**Problem:**
- Co-adaptation: Neurons develop dependency
- Overfitting: Model memorizes training data

**How Dropout Works:**
```
During training:
- Randomly disable (drop) fraction p of neurons
- Remaining network must learn robust features
- Like training ensemble of thinned networks

During inference:
- Use all neurons
- Scale by (1 - p) to account for all active neurons
```

**Code:**
```python
model = Sequential([
    Dense(128, activation='relu'),
    Dropout(0.5),         # Drop 50% of neurons
    Dense(64, activation='relu'),
    Dropout(0.3),         # Drop 30%
    Dense(10, activation='softmax')
])
```

**Dropout Impact:**
- Reduces overfitting
- Requires more epochs to converge
- Acts as averaging over models
- No cost during inference

**Typical Values:**
```
Hidden layers: 0.3 - 0.5 (30-50% dropout)
Input layer: 0.1 - 0.2 (low dropout)
Output layer: No dropout
```

**Comparison of Techniques:**

| Technique | Purpose | Cost |
|-----------|---------|------|
| Learning Rate Scheduling | Optimize convergence | Computational |
| Batch Normalization | Stability + Regularization | Slight overhead |
| Dropout | Prevent overfitting | 20-30% slower training |

**Combined Typical Setup:**

   - Unfreeze later layers
   - Train with low learning rate
   - Balance between pre-trained knowledge and adaptation
   - Most common approach

3. **Full Retraining** (All Unfrozen)
   - Train entire network

### Q9: Explain vanishing and exploding gradients and solutions.

**Answer:**

**Vanishing Gradient Problem**

**What Happens:**
```
Loss = (predicted - actual)²
∂L/∂W3 = (chain of multiplications)
∂L/∂W1 = (even longer chain)

If each term < 1: 0.9 × 0.9 × 0.9 × ... →near 0
Result: Early layers don't update (don't learn)
```

**Visual:**
```python
# Sigmoid derivative
σ'(x) = σ(x) × (1 - σ(x))  # Always < 0.25

# Deep network with 100 layers
gradient = 0.25^100 ≈ 0 (essentially zero!)
```

**Impact:**
- Network can't learn long-range dependencies
- RNNs especially affected
- Training stalls

**Solutions:**

1. **Use ReLU Instead of Sigmoid**
   ```
   ReLU derivative = 1 (when x > 0)
   vs
   Sigmoid derivative = 0.25 (max)
   ```

2. **Batch Normalization**
   ```python
   from tensorflow.keras.layers import BatchNormalization
   
   model.add(Dense(128))
   model.add(BatchNormalization())
   model.add(Activation('relu'))
   
   # Normalizes inputs to each layer
   # Keeps activations in good range
   ```

3. **Skip Connections (Residual Networks)**
   ```
   Normal: x → Conv → Conv → output
   ResNet: x → Conv → Conv ⸻(add)⸻ output
                                 ↑___x___↑
   
   Allows gradient to flow directly
   ```

4. **LSTM/GRU for RNNs**
   ```python
   # Instead of RNN
   from tensorflow.keras.layers import LSTM
   
   # LSTM has gating mechanism
   # Explicitly controls gradient flow
   model.add(LSTM(128))
   ```

**Exploding Gradient Problem**

**What Happens:**
```
Similar chain multiplication, but terms > 1
0.2 → 2 → 2 → ... (exponentially growing)
Result: NaN values, weights become huge
```

**Solutions:**

1. **Gradient Clipping**
   ```python

       loss='mse',
       metrics=['mae'],
       clipnorm=1.0  # Max norm = 1
   )
   
   # Clips gradients if norm exceeds threshold
   ```

2. **Weight Normalization**
   ```python
   # Normalize weight matrices
   # Prevents them from becoming too large
   ```

3. **Learning Rate Scheduling**
   ```python
   # Reduce learning rate over time
   lr_schedule = exponential_decay(0.1, decay_steps=1000)
   ```

### Q10: What is attention mechanism and why is it revolutionary?

**Answer:**

**Attention Mechanism** - Allows model to focus on relevant parts of input dynamically.

**Problem it Solves:**
```
Machine Translation: "The bank can lend money"

Without Attention:
- Encode entire sentence into single vector
- Lose information about which word matters where
- Fixed representation regardless of context

With Attention:
- When translating "bank" → can focus on "lend"
- When translating "lend" → can focus on "bank"
- Dynamic focus on relevant context
```

**How It Works:**

```
For each position, compute:
1. Query (current position): "What am I looking for?"
2. Key (all positions): "What can I offer?"
3. Value (all positions): "Here's what I have"

Attention Score = softmax(Query ⊙ Key^T / √d_k)
Output = Attention Score × Value

Example:
Position=2 (word "bank"):
- Query2 = learned representation of current position
- Key1...Key5 = representations of all words
- Scores = [0.1, 0.8, 0.05, 0.02, 0.03]  (highest for word "lend")
- Output2 = weighted sum using these scores
```

**Self-Attention:**
```
Query, Key, Value all from same input
"Bank" attends to all words including itself
Captures relationships between all word pairs
```

**Multi-Head Attention:**
```
Multiple attention operations in parallel
Each head focuses on different aspects of relationships
Results concatenated

Benefits:
- Richer representation
- Capture diverse patterns
- Standard in transformers (8-12 heads typical)
```

**Why Revolutionary:**

1. **Parallel Processing**
   - RNNs process sequentially (slow for long sequences)
   - Attention is parallel (fast on GPUs)
   - Enables training on huge datasets

2. **Long-Range Dependencies**
   - Every position can attentively relate to every other
   - No information bottleneck like in RNNs

3. **Interpretability**
   - Attention weights show what model focuses on
   - Can visualize which words matter

4. **Performance**
   - BERT, GPT, T5, Claude all based on attention
   - State-of-the-art results in NLP

**Code (Simplified):**
```python
import tensorflow as tf

class AttentionLayer(tf.keras.layers.Layer):
    def call(self, query, key, value):
        # Compute attention scores
        scores = tf.matmul(query, key, transpose_b=True)
        scores = scores / tf.math.sqrt(tf.cast(tf.shape(key)[-1], tf.float32))
        
        # Apply softmax
        weights = tf.nn.softmax(scores, axis=-1)
        
        # Weight values
        output = tf.matmul(weights, value)
        return output
```

### Q11: Explain reinforcement learning concepts and applications.

**Answer:**

**Reinforcement Learning (RL)** - Agent learns by interacting with environment, receiving rewards/penalties.

**Key Components:**

1. **Agent** - The learner (AI system)
2. **Environment** - The world
3. **State** - Current situation
4. **Action** - What agent can do
5. **Reward** - Feedback (+/-)
6. **Policy** - Strategy for choosing actions

**Process:**
```
repeat:
    1. Observe state
    2. Choose action (explore or exploit?)
    3. Receive reward
    4. Update knowledge about state-action value
until convergence
```

**Key Algorithms:**

1. **Q-Learning (Value-Based)**
   ```python
   # Learn Q(state, action) = expected future reward
   Q[state][action] += learning_rate * (reward + max(Q[next_state]) - Q[state][action])
   
   # Choose action: Pick action with highest Q value
   action = argmax(Q[state])
   ```

2. **Policy Gradient**
   ```
   Directly learn policy (probability of actions)
   Update policy to increase probability of rewarding actions
   Used in: Actor-Critic, PPO, A3C
   ```

3. **Actor-Critic**
   ```
   Two networks:
   - Actor: Choose actions (policy)
   - Critic: Evaluate actions (value function)
   More stable than pure policy gradient
   ```

**Exploration vs Exploitation:**
```
Exploration: Try new actions (might find better strategy)
Exploitation: Use known best action (maximize immediate reward)

Epsilon-Greedy:
- With probability ε: explore (random action)
- With probability 1-ε: exploit (best known action)
```

**Applications:**

1. **Game Playing**
   - AlphaGo beats world champion
   - Dota 2, Chess AI

2. **Robotics**
   - Robot learns to walk/grasp
   - Policies transfer between tasks

3. **Autonomous Driving**
   - Agent learns driving policy
   - Avoid obstacles, reach goal

4. **Resource Allocation**
   - Network optimization
   - Data center management

5. **Trading**
   - Portfolio optimization
   - Buy/sell decisions

**Challenges:**
- Sample inefficiency (needs many interactions)
- Sparse rewards (rare feedback)
- Unstable training
- Sim-to-real transfer (simulation vs reality gap)

### Q3: What is backpropagation and how does it work?

**Answer:**

**Backpropagation** - Algorithm to compute gradients and update network weights to minimize error.

**Three Phases:**

1. **Forward Pass**
   - Data moves through network
   - Compute predictions
   - Calculate error (loss)

2. **Backward Pass (Backpropagation)**
   - Compute gradient of loss with respect to each weight
   - Use chain rule: Derivative flows backward through layers
   - Determine how much each weight contributed to error

3. **Update Weights**
   - Adjust weights in direction that reduces error
   - Learning rate controls step size
   - Formula: W_new = W_old - learning_rate * gradient

**Mathematical Intuition:**
```
Loss Function: L = (predicted - actual)²

For third layer: ∂L/∂W3 = ∂L/∂output × ∂output/∂W3
For second layer: ∂L/∂W2 = ∂L/∂W3 × ∂W3/∂input2 × ∂input2/∂W2
...and so on, flowing backward
```

**Visualization:**
```
Forward Pass:
x → [W1] → h1 → [W2] → h2 → [W3] → ŷ
                                      ↓
                                  Loss = L

Backward Pass:
∂L/∂W3 ← ∂L/∂ŷ ← Loss
∂L/∂W2 ← ∂L/∂h2 ← ∂L/∂W3
∂L/∂W1 ← ∂L/∂h1 ← ∂L/∂W2

Weight Update:
W3 = W3 - α × ∂L/∂W3
W2 = W2 - α × ∂L/∂W2
W1 = W1 - α × ∂L/∂W1  (α = learning rate)
```

**Backprop Algorithm Summary:**
```python
for epoch in range(num_epochs):
    # Forward pass
    for batch in batches:
        predictions = forward(X_batch)
        loss = compute_loss(predictions, y_batch)
        
        # Backward pass
        gradients = compute_gradients(loss)
        
        # Update weights
        weights = weights - learning_rate * gradients
```

### Q5: What are activation functions and when to use each?

**Answer:**

**Activation Function** - Non-linear transformation applied to neuron outputs. Enables network to learn non-linear patterns.

**Common Activation Functions:**

1. **ReLU (Rectified Linear Unit)**
   ```
   f(x) = max(0, x)
   Returns x if positive, 0 if negative
   
   Pros:
   - Simple, fast computation
   - Avoids vanishing gradient
   - Default choice for hidden layers
   
   Cons:
   - Dead neurons (ReLU=0 forever)
   - Not differentiable at x=0
   
   When: Most hidden layers (default)
   ```

2. **Sigmoid**
   ```
   f(x) = 1 / (1 + e^(-x))
   Returns value between 0 and 1
   
   Pros:
   - Interpretable as probability
   - Differentiable everywhere
   
   Cons:
   - Vanishing gradient problem
   - Computationally expensive
   
   When: Binary classification (last layer)
   ```

3. **Tanh (Hyperbolic Tangent)**
   ```
   f(x) = (e^x - e^(-x)) / (e^x + e^(-x))
   Returns value between -1 and 1
   
   Pros:
   - Zero-centered (helps training)
   - Differentiable
   
   Cons:
   - Still suffers from vanishing gradient
   
   When: RNN hidden layers, when zero-center is important
   ```

4. **Softmax**
   ```
   f(x_i) = e^x_i / Σe^x_j
   Returns probabilities summing to 1
   
   Pros:
   - Multi-class probability distribution
   - Interpretable
   
   Cons:
   - Only for output layer
   
   When: Multi-class classification output layer
   ```

5. **Leaky ReLU**
   ```
   f(x) = x if x > 0, else 0.01x
   Small negative slope prevents dead neurons
   
   Pros:
   - Prevents dead neurons
   - Smoother gradient
   
   When: When basic ReLU causes dead neurons
   ```

**Decision Guide:**
```
Hidden Layers:
- Default: ReLU
- Problematic: Leaky ReLU or ELU
- RNN: Tanh

Output Layer:
- Binary classification: Sigmoid
- Multi-class: Softmax
- Regression: Linear (no activation)
```

### Q6: Explain gradient descent and learning rate.

**Answer:**

**Gradient Descent** - Optimization algorithm to minimize loss by iteratively updating parameters.

**How it Works:**
```
1. Calculate gradient (slope) of loss function
2. Move in opposite direction (downhill)
3. Step size = learning_rate × gradient
4. Repeat until convergence
```

**Intuition - Walking Down a Hill:**
```
You're on a hill blindfolded
Can feel the slope under your feet (gradient)
Take steps downhill (negative gradient direction)
Bigger steps with steep slope (large gradient)
Smaller steps with gentle slope (small gradient)
Eventually reach bottom (minimum loss)
```

**Learning Rate** - Controls step size.

```python
W_new = W_old - learning_rate × gradient

# Example
W_old = 5
gradient = 0.5

# High learning rate (0.5)
W_new = 5 - 0.5 × 0.5 = 4.75  # Large step

# Low learning rate (0.01)
W_new = 5 - 0.01 × 0.5 = 4.995  # Small step
```

**Learning Rate Impact:**

Too High:
```
Loss keeps oscillating or diverging
May overshoot minimum
Never converges
```

Too Low:
```
Very slow convergence
Takes forever to train
Gets stuck in local minimums
```

Optimal:
```
Smooth convergence to minimum
Reaches goal in reasonable time
```

**Variants:**

1. **Batch Gradient Descent**
   - Compute gradient on entire dataset
   - Stable but slow for large data

2. **Stochastic Gradient Descent (SGD)**
   - Update on single sample
   - Fast but noisy

3. **Mini-batch Gradient Descent**
   - Update on small batches
   - Balance between stability and speed (typical approach)

4. **Adaptive Methods**
   ```python
   # Adam (Adaptive Moment Estimation)
   Combines momentum and RMSprop
   Adjusts learning rate per parameter
   Usually converges faster than SGD
   ```

**Code:**
```python
# TensorFlow/Keras uses optimizers internally
model.compile(optimizer='adam', loss='mse')
# 'adam' automatically handles learning rate

# But you can control it
from tensorflow.keras.optimizers import Adam
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
```

### Q7: What are common challenges when training deep neural networks?

**Answer:**

**Challenge 1: Vanishing Gradient Problem**
```
Problem:
- During backpropagation, gradients get multiplied many times
- In deep networks, gradients become tiny (vanish)
- Parameters in early layers don't update

Solution:
- Use ReLU instead of Sigmoid/Tanh
- Batch normalization
- Skip connections (ResNet)
- Use LSTM/GRU for RNNs
```

**Challenge 2: Exploding Gradient**
```
Problem:
- Opposite of vanishing
- Gradients become very large
- Unstable training, NaN values

Solution:
- Gradient clipping: Max(gradient, threshold)
- Careful weight initialization
- Batch normalization
```

**Challenge 3: Overfitting**
```
Problem:
- Model memorizes training data
- Poor generalization to test data

Solution:
- Dropout: Randomly disable neurons during training
- Data augmentation: Create more training samples
- Regularization (L1/L2): Penalize large weights
- Early stopping: Stop when validation loss increases
- Use less complex model
```

**Challenge 4: Computational Cost**
```
Problem:
- Deep networks require significant GPU/CPU
- Training takes days/weeks

Solution:
- Use transfer learning (pretrained models)
- Distributed training (multiple GPUs)
- Model compression techniques
- Batch processing efficiency
```

**Challenge 5: Dead Neurons (ReLU)**
```
Problem:
- ReLU(x) = 0 for negative x
- Sometimes neuron outputs always negative
- Stays dead forever (no gradient updates)

Solution:
- Use Leaky ReLU: small negative slope
- Use ELU activation
- Xavier initialization
```

**Challenge 6: Poor Hyperparameter Choices**
```
Problem:
- Wrong learning rate, batch size, number of layers
- No convergence or overfitting

Solution:
- Use grid search or Bayesian optimization
- Start with defaults from literature
- Monitor training/validation loss plots
- Use learning rate scheduling
```

**Key Characteristics:**
- Reduces bias and variance
- Sequential (not parallel)
- Takes longer but often more accurate
- Examples: AdaBoost, Gradient Boosting, XGBoost

**Example:**
```
Base Learners:
- Logistic Regression
- Random Forest
- SVM
  ↓ (their predictions become features)
Meta-Learner:
- Logistic Regression (learns how to combine)
  ↓
Final Prediction
```

### Q14: Explain data leakage and how to prevent it.

**Answer:**

**Data Leakage** - Information from test/validation data leaks into training, causing overly optimistic results.

**Types of Leakage:**

1. **Information from Future**
   ```python
   # WRONG: Using future information
   df = df.sort_values('date')
   df['future_price'] = df['price'].shift(-1)
   
   # Model uses future price to predict current
   # Real prediction: Can't access future!
   ```

2. **Target Leakage**
   ```python
   # WRONG: Including target-related info as feature
   df['fraud_was_caught'] = fraud_detection_result
   df['investigation_conducted'] = target == 1  # LEAKAGE!
   
   # Correct: Use only pre-incident information
   df['account_age'] = now - account_creation_date
   ```

3. **Temporal Leakage**
   ```python
   # WRONG: Normalize entire dataset
   scaler = StandardScaler()
   X_scaled = scaler.fit_transform(X)  # Uses ALL data statistics
   split_point = int(0.8 * len(X))
   X_train = X_scaled[:split_point]
   X_test = X_scaled[split_point:]
   
   # Correct: Fit scaler on train only
   scaler.fit(X_train)
   X_train_scaled = scaler.transform(X_train)
   X_test_scaled = scaler.transform(X_test)
   ```

4. **Preprocessing Leakage**
   ```python
   # WRONG: Impute missing values before split
   df['age'].fillna(df['age'].mean(), inplace=True)
   X_train = df[:-100]
   X_test = df[-100:]
   
   # Correct: Impute within cross-validation
   from sklearn.pipeline import Pipeline
   from sklearn.compose import ColumnTransformer
   
   pipeline = Pipeline([
       ('imputer', SimpleImputer(strategy='mean')),
       ('model', LogisticRegression())
   ])
   
   cv_scores = cross_val_score(pipeline, X, y, cv=5)
   ```

5. **Feature Selection Leakage**
   ```python
   # WRONG: Select features on entire dataset
   selected_features = chi2_select_features(X, y, k=10)
   X_train = X_train[selected_features]
   X_test = X_test[selected_features]
   
   # Correct: Select within pipeline
   pipeline = Pipeline([
       ('feature_selection', SelectKBest(k=10)),
       ('model', LogisticRegression())
   ])
   ```

**Prevention Checklist:**
```
✅ Split data FIRST before any preprocessing
✅ Fit preprocessors on training data only
✅ Apply fitted preprocessor to test data
✅ Never use test data statistics in training
✅ Use time-based split for time series
✅ Carefully review features for information not available at prediction time
✅ Validate with test set separate from validation used during tuning
```

### Q9: What are evaluation metrics for classification and regression?

**Answer:**

**Classification Metrics:**

1. **Accuracy**
   ```
   = (TP + TN) / Total
   When to use: Balanced dataset
   Problem: May be misleading with imbalanced data
   ```

2. **Precision & Recall**
   ```
   Precision = TP / (TP + FP)  → False positives cost high
   Recall = TP / (TP + FN)     → False negatives cost high
   
   Example: Medical diagnosis
   - High precision: Avoid false alarms (don't stress patient)
   - High recall: Don't miss actual cases (safety critical)
   ```

3. **F1-Score**
   ```
   Harmonic mean of precision and recall
   F1 = 2 * (Precision * Recall) / (Precision + Recall)
   Good for imbalanced datasets
   ```

4. **ROC-AUC (Receiver Operating Characteristic - Area Under Curve)**
   ```
   - Plots true positive rate vs false positive rate
   - AUC = 1: Perfect classifier
   - AUC = 0.5: Random guessing
   - Good with imbalanced data
   ```

5. **Confusion Matrix**
   ```
   |         | Predicted Yes | Predicted No |
   |---------|---------------|--------------|
   | Actual Yes | TP          | FN           |
   | Actual No  | FP          | TN           |
   ```

**Regression Metrics:**

1. **Mean Absolute Error (MAE)**
   ```
   MAE = Σ|Yi - Ŷi| / n
   Robust to outliers
   Example: $500 average error
   ```

2. **Mean Squared Error (MSE)**
   ```
   MSE = Σ(Yi - Ŷi)² / n
   Sensitive to outliers
   Squared units harder to interpret
   ```

3. **Root Mean Squared Error (RMSE)**
   ```
   RMSE = √MSE
   Same units as target
   Commonly used
   ```

4. **R² Score**
   ```
   R² = 1 - (SS_res / SS_tot)
   Range: [-∞, 1]
   1: Perfect fit
   0: Model as good as mean
   Negative: Worse than mean
   ```

5. **Mean Absolute Percentage Error (MAPE)**
   ```
   MAPE = (1/n) * Σ|Yi - Ŷi| / |Yi| * 100
   Good for percentage-based comparison
   ```

**Choosing the Right Metric:**

| Scenario | Metric |
|----------|--------|
| Spam detection | Precision (false positives are bad) |
| Disease screening | Recall (false negatives are bad) |
| Imbalanced data | F1-Score, AUC-ROC |
| Regression (no outliers) | RMSE |
| Regression (outliers present) | MAE |

### Q4: Explain key statistical concepts: Mean, Median, Mode, Standard Deviation

**Answer:**

**Mean (Average)**
```
Sum of all values / Number of values
Example: [1, 2, 3, 4, 5] → Mean = 15/5 = 3
Affected by outliers
```

**Median (Middle Value)**
```
Middle value when sorted
For even count: Average of two middle values
Example: [1, 3, 7, 9] → Median = (3+7)/2 = 5
Robust to outliers
```

**Mode (Most Frequent)**
```
Most frequently occurring value
Example: [1, 2, 2, 3, 2] → Mode = 2
Can have multiple modes
```

**Standard Deviation (Spread)**
```
Measures how spread out data is from mean
σ = √(Σ(Xi - Mean)²/N)
Example: σ=0 means all identical, high σ means very spread out
68% data within ±1σ (normal distribution)
```

**Choosing the Right Metric:**

| Metric | Best When | Example |
|--------|-----------|---------|
| Mean | Symmetric distribution, no outliers | Temperature readings |
| Median | Skewed distribution, outliers exist | Household income |
| Mode | Categorical data | Favorite color |
| Std Dev | Understanding data spread | Risk assessment |

**Code Example:**
```python
import numpy as np

data = [1, 2, 2, 3, 100]  # Outlier present
print(np.mean(data))      # 21.6 (affected by outlier)
print(np.median(data))    # 2.0 (not affected)
print(np.std(data))       # 38.5 (high due to outlier)
```

- Distance-based algorithms biased
- Optimization slower

✅ **Benefit with scaling:**

- Better model performance

### Q13: Design a distributed rate limiter for an API

**Answer:**

**Requirements:**
- Limit requests per user/IP per time window
- Distributed across multiple servers
- Sub-millisecond lookup
- Support multiple strategies

**Strategies:**

1. **Token Bucket Algorithm**
   ```
   - Bucket has capacity C tokens
   - Refill R tokens per second
   - Each request costs 1 token
   - Request allowed if tokens available
   ```
   - Pro: Handles bursts smoothly
   - Con: Requires refill calculation

2. **Sliding Window with Counter**
   ```
   - Track request counts in Redis
   - Sliding window ensures precise limit
   - More accurate than fixed windows
   ```

3. **Leaky Bucket Algorithm**
   ```
   - Requests added to queue
   - Leak at constant rate
   - Smooth traffic shaping
   ```

**Implementation Architecture:**

```python
class DistributedRateLimiter:
    def __init__(self, redis_client):
        self.redis = redis_client
    
    def is_allowed(self, user_id, limit=100, window=60):
        key = f"rate_limit:{user_id}"
        current_count = self.redis.get(key) or 0
        
        if current_count >= limit:
            return False
        
        pipe = self.redis.pipeline()
        pipe.incr(key)
        pipe.expire(key, window)
        pipe.execute()
        return True
```

**Distributed Considerations:**
- **Redis Clustering** for high availability
- **Sticky Sessions** to same rate limiter node
- **Atomic Operations** to prevent race conditions
- **Fallback Strategy** if Redis unavailable

### Q10: What is the CAP theorem and its implications for database design?

**Answer:**

**CAP Theorem** states distributed systems can guarantee only 2 of 3 properties:

1. **Consistency (C)**
   - All nodes see same data simultaneously
   - No stale data
   - Requires coordination

2. **Availability (A)**
   - System always operational
   - Request never fails (within timeout)
   - Requires no coordination

3. **Partition Tolerance (P)**
   - System works despite network partitions
   - Network messages may be lost
   - Almost always required in distributed systems

**CAP Trade-offs:**

```
Choose 2:
- CP (Consistency + Partition): PostgreSQL, MongoDB (strict mode)
  → Accept unavailability during partitions
  → Example: Banking systems
  
- AP (Availability + Partition): Memcached, Cassandra
  → Accept eventual consistency
  → Example: Social media feeds
  
- CA (Consistency + Availability): Single-node databases
  → Not partition tolerant
  → Example: Monolithic databases
```

**Practical Implications:**

1. **System Design Decision**
   - Financial: Prioritize consistency
   - Social media: Prioritize availability
   - E-commerce: Mix based on feature

2. **Eventual Consistency**
   - Accept temporary inconsistency
   - Consistency achieved over time
   - Works for most web applications

3. **Conflict Resolution**
   - Last-write-wins
   - Vector clocks
   - Application-level merging

**Real-World Examples:**
- **Google Bigtable**: Favors CP
- **Amazon DynamoDB**: Favors AP
- **PostgreSQL**: Favors CP

### Q1: Explain the SOLID principles and their importance in software design.

**Answer:**
SOLID is an acronym representing five design principles that promote clean, maintainable code:

1. **S - Single Responsibility Principle (SRP)**
   - Each class should have only one reason to change
   - Example: A `UserService` handles user logic, NOT database operations
   
2. **O - Open/Closed Principle**
   - Classes should be open for extension, closed for modification
   - Use inheritance/interfaces instead of modifying existing code
   
3. **L - Liskov Substitution Principle**
   - Objects of a superclass should be replaceable with objects of its subclasses
   - Derived classes must extend functionality, not narrow it
   
4. **I - Interface Segregation Principle**
   - Clients should not depend on interfaces they don't use
   - Many client-specific interfaces better than one general-purpose interface
   
5. **D - Dependency Inversion Principle**
   - High-level modules should not depend on low-level modules
   - Both should depend on abstractions

**Why it matters:** Reduces coupling, improves testability, makes code reusable and maintainable.

| **Underfetching** | Common (need multiple calls) | Rare (all data in one call) |
| **Caching** | Built-in HTTP caching | Requires custom caching strategies |
| **Learning Curve** | Easier to learn | Steeper learning curve |
| **Best For** | Simple, standard operations | Complex, interconnected data |

### Q3: What is an OOP concept, and what are the four pillars?

**Answer:**

**OOP (Object-Oriented Programming)** organizes code around "objects" that combine data (attributes) and behavior (methods).

**Four Pillars:**

1. **Encapsulation**
   - Bundle data and methods together
   - Hide internal implementation (private/protected members)
   - Expose only necessary public interface
   - Benefit: Security and maintainability

2. **Abstraction**
   - Show only essential features, hide complexity
   - Abstract classes and interfaces define contracts
   - User doesn't need to know internal workings
   - Benefit: Simplified interfaces

3. **Inheritance**
   - Create hierarchy of classes (parent/child)
   - Child inherits properties/methods from parent
   - Promotes code reuse
   - Benefit: Reusability

4. **Polymorphism**
   - Same method works with different object types
   - Method overriding (runtime) and overloading (compile-time)
   - Enables flexibility and extensibility
   - Benefit: Write generic code

### Q4: What is the difference between composition and inheritance?

**Answer:**

| Aspect | Inheritance | Composition |
|--------|-------------|------------|
| **Relationship** | "IS-A" (Car is a Vehicle) | "HAS-A" (Car has an Engine) |
| **Syntax** | `class Car extends Vehicle` | `Car has property Engine` |
| **Flexibility** | Less flexible, tight coupling | More flexible, loose coupling |
| **Reuse** | Code reuse through hierarchy | Reuse through object aggregation |
| **Maintenance** | Changes to parent affect all children | Changes isolated to components |
| **Best For** | Natural hierarchy, shared behavior | Avoiding tight coupling |

**Example:**
```python
# Inheritance (Less Flexible)
class Animal:
    def eat(self): pass

class Dog(Animal):
    pass

# Composition (More Flexible)
class Engine:
    def start(self): pass

class Car:
    def __init__(self):
        self.engine = Engine()
    def start(self):
        self.engine.start()
```

**When to use Composition:** Composition is generally preferred in modern design because it's more flexible and avoids fragile base class problems.

### Q5: Explain the MVC (Model-View-Controller) architecture pattern.

**Answer:**

**MVC** separates application into three interconnected components:

1. **Model**
   - Represents data and business logic
   - Independent of UI
   - Handles data validation and database operations
   - Example: User database model

2. **View**
   - Presents data to the user
   - Receives user input
   - Read-only access to Model (displays data only)
   - Example: HTML/CSS UI

3. **Controller**
   - Handles user input from View
   - Updates Model based on user actions
   - Determines which View to display
   - Acts as liaison between Model and View
   - Example: Processing form submissions

**Flow:**
```
User Input → Controller → Model (update data) → View (display updated data)
```

**Advantages:**
- Separation of concerns
- Easy to test (Model can be tested independently)
- Supports multiple views for same Model
- Facilitates parallel development

### Q6: What is a design pattern? Name 3 common creational patterns.

**Answer:**

**Design Pattern** is a reusable solution to common programming problems. It's a template for solving a problem in a way that's proven and can be reused.

**Common Creational Patterns:**

1. **Singleton Pattern**
   - Ensures only one instance of a class exists
   - Provides global access point
   - Example: Database connection pool
   ```python
   class Database:
       _instance = None
       def __new__(cls):
           if cls._instance is None:
               cls._instance = super().__new__(cls)
           return cls._instance
   ```

2. **Factory Pattern**
   - Creates objects without exposing creation logic
   - Encapsulates object instantiation
   - Example: Creating different document types
   ```python
   class DocumentFactory:
       @staticmethod
       def create(doc_type):
           if doc_type == 'pdf':
               return PDFDocument()
           elif doc_type == 'word':
               return WordDocument()
   ```

3. **Builder Pattern**
   - Constructs complex objects step-by-step
   - Separates construction from representation
   - Example: Building a complex database query
   ```python
   class QueryBuilder:
       def __init__(self):
           self.query = ""
       def select(self, columns):
           self.query += f"SELECT {columns} "
           return self
       def where(self, condition):
           self.query += f"WHERE {condition}"
           return self
   ```
